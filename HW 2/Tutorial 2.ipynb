```python
import numpy as np
import matplotlib.pyplot as plt
import rk4 #rk4(y0, x0, f, h, n, f_args = {})
```

```python
y0 = np.array([1])
x0 = 0
h = 1
n = 1000
r=1

def f(y, x):
    return -r*y
yn, xn = rk4.rk4(y0,x0,f,h,n)
```

```python
plt.figure(figsize=(8,5))
plt.title("Solution of $\dot{x} = -rx$ for $r=1$")
plt.xlabel("x")
plt.xlim((0,100))
plt.yscale("log")
plt.ylabel("y")
plt.ylim(1e-20,1e-2)
plt.plot(xn, np.exp(-r*xn), label="Analytical solution")
plt.plot(xn,yn, label="Integrated solution")
plt.legend()
plt.show()
```
